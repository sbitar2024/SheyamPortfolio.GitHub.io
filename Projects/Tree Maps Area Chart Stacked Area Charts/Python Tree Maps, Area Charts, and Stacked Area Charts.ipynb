```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify
import seaborn as sns


data = pd.read_csv('unemployement-rate-1948-2010.csv')

plt.figure(figsize=(10, 8))
squarify.plot(sizes=data.groupby('Year')['Value'].sum(), label=data['Year'].unique(), alpha=0.7)
plt.axis('off')
plt.title('Python Unemployment Rate by Year (1948-2010) - Treemap')
plt.show()
```

```python
#Area chart

plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Year', y='Value', estimator=sum, ci=None)
plt.title('Python Unemployment Rate Over Years (1948-2010) - Area Chart')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show() 
```

```python
#Stacked area chart
import matplotlib.pyplot as plt

#Extracting years and unemployment rates
years = data['Year']
unemployment_rates = data['Value']

#Creating stacked area chart
plt.figure(figsize=(10, 6))
plt.stackplot(years, unemployment_rates, labels=['Unemployment Rate'], alpha=0.5)
plt.title('Python Unemployment Rate Over Time - Stacked area chart')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```
