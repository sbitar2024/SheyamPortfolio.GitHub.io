Python: Histograms, Box Plots, & Bullet Charts



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
# Read the data from the CSV file
df = pd.read_csv("education.csv")

# Drop the row where state is "United States"
df = df[df['state'] != 'United States']

df
```

```python
#Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['reading'], bins=15, color='skyblue', edgecolor='black')
plt.title('Python Histogram of Reading Scores')
plt.xlabel('Reading Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['reading', 'math', 'writing']])
plt.title('Python Box Plot of Reading, Math, and Writing Scores')
plt.ylabel('Score')
plt.show()

```

```python
# Bullet Chart
plt.figure(figsize=(10, 10))
sns.scatterplot(data=df, x='percent_graduates_sat', y='state', hue='state', palette='viridis', legend=False)
plt.title('Python Bullet Chart of Percent Graduates SAT')
plt.xlabel('Percent Graduates SAT')
plt.ylabel('State')
plt.grid(True)
plt.show()

```

```python
# Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='dropout_rate', data=df, inner='point', color='purple')
plt.title('Python Violin Plot of Dropout Rate')
plt.xlabel('Dropout Rate')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
